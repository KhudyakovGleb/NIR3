```python
df_nspb = pd.read_csv(r'd:\Downloads\problems_export_2024-06-06.csv')
df_nspb['Подкатегория'].unique()
```

```python
df_nspb = df_nspb[(df_nspb['Подкатегория'] == 'Загрязнение водного объекта или его берега')]
df_nspb[(df_nspb['Подкатегория'] == 'Загрязнение водного объекта или его берега')]
df_nspb['Подкатегория'] = 'Экология'
df_nspb = df_nspb[['Описание', 'Подкатегория']]
df_nspb = df_nspb.rename(columns={'Описание': 'Текст', 'Подкатегория': 'Блок по распределению'})
df_POS = df_POS[['Текст', 'Блок по распределению']]
df = pd.concat([df_POS, df_nspb], ignore_index=True)
df
```

```python
import re
def clean_text(text):

    text = text.lower()
    text = re.sub(r'[^\w\s]', '', text)  
    text = re.sub(r'[A-Za-z]', '', text)
    text = re.sub(r'\d+', '', text) 
    text = re.sub(r'\n', ' ', text)
    return text

df = df.dropna()
df['Текст'] = df['Текст'].apply(clean_text)
df = df[df['Текст'].str.len() >= 5]
```

```python
df = df.groupby('Блок по распределению', group_keys=False).apply(lambda x: x.sample(frac=1)).groupby('Блок по распределению').head(2000)
df
```

```python
len_cat = {}
for cats in df['Блок по распределению'].unique():
    len_cat[cats] = len(df[(df['Блок по распределению'] == cats)])

print(len_cat)
```

```python
df['Блок по распределению'] = df['Блок по распределению'].replace({'ТКО':'Отходы', 'БОС': 'Безопасность'})
```

```python
df
```

```python
dic_lables = {}
counter = 0
for cats in df['Блок по распределению'].unique():
    dic_lables[counter] = cats
    counter += 1
dic_lables
```

```python
with open("dic_lables.txt", "w") as file:
    for key, value in dic_lables.items():
        file.write(f"{key}: {value}\n")
```

```python
for key, value in dic_lables.items():
    df['Блок по распределению'] = df['Блок по распределению'].replace({value:key})
```
